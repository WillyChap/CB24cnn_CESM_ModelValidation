```python
clevels = [0.5,0.6,.7,.8,.9,1,1.1,1.15,1.2,1.25,1.3,1.35,1.4,1.45,1.5,1.6]
Bebop.plot_background_removed(labels=True,figsize=[15,10],contour_range=(0.5,2,0.1),text_size=18,clevels=clevels)
# Bebop.save_figs(name='f.e.FTORCHmjo_ERA5')
```

```python
psumanti_r = Bebop.wk_spectra['psumanti_r']
psumsym_r = Bebop.wk_spectra['psumsym_r']
wavefft = Bebop.wk_spectra['wavefft']
freqfft = Bebop.wk_spectra['freqfft']
indwave = np.where(np.logical_and(wavefft>=-Bebop.max_wn_plot-0.001,wavefft<=Bebop.max_wn_plot+0.001))[0]
indfreq = np.where(np.logical_and(freqfft>0,freqfft<=Bebop.max_freq_plot+0.1))[0]
waveplt = wavefft[indwave]
freqplt = freqfft[indfreq]
powerplt = psumsym_r[indfreq,:]
powerplt = powerplt[:,indwave]
Y,X = np.meshgrid(freqplt,waveplt)

##### plotting: #####################
figAS, axAS = plt.subplots(1,1,figsize=(10,8))

axAS.tick_params(labelsize=13, direction='out') 
axAS.grid(False, alpha=0.9)

# Set x-axis ticks
x_ticks = np.linspace(0.006, 0.05, 10)  # Adjust the number and position of ticks
axAS.set_xticks(x_ticks)
axAS.set_xticklabels([f'{x:.3f}' for x in x_ticks], fontsize=13)  # Formatting tick labels to 3 decimal places

# Set y-axis ticks
y_ticks = np.linspace(0, 6, 7)  # Adjust the number and position of ticks
axAS.set_yticks(y_ticks)
axAS.set_yticklabels([int(y) for y in y_ticks], fontsize=13)

cpd_lines = [80,30]
for d in cpd_lines:
        axAS.axvline(1./d,color='k', linestyle='--')
        axAS.text(1./d,(1./d+5),str(d)+' days',\
        size=16,bbox={'facecolor':'white','alpha':0.9,\
        'edgecolor':'none'})


cpd_lines = [80,70,60,50,40,30,20,10]
for d in cpd_lines:
    axAS.axvline(1./d,color='k', linestyle='--',alpha=0.2)

cset_1  = axAS.contourf(Y,X,powerplt.T,extend='both',levels=np.arange(0,2.95,.15),cmap='gist_ncar',alpha=0.9)

plt.ylim([0,6])
plt.xlim([0.006,0.05])
axAS.grid(False, alpha=0.9)
axAS.set_xlabel('cycles per day', fontsize=13)
axAS.set_ylabel('Zonal Wave Number', fontsize=13)
plt.savefig(save_Fig_out_n,bbox_inches='tight',dpi=400)
plt.grid(False)
cbar = plt.colorbar(cset_1)
cbar.ax.tick_params(labelsize=12)
plt.show()
```
