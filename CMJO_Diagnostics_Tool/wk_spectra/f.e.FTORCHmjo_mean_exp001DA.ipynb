```python
import wk_spectra.wk_analysis as wka
import warnings
warnings.filterwarnings("ignore")
```

```python
Bebop = wka.wk_analysis()
filep = '/glade/derecho/scratch/wchapman/ADF/f.e.FTORCHmjo_mean_exp001DA/ts/'
filen = 'f.e.FTORCHmjo_mean_exp001DA.h1.FLUT.1979010100000-2000122200000.nc'
filenn = filep +filen
Bebop.import_netcdf(file=filenn,varname='FLUT',latname='lat',latBound=15)
Bebop.wheeler_kiladis_spectra(spd=1,nDayWin=96,nDaySkip=30,max_freq=0.5,max_wn=20)
Bebop.print_status()
```
